```python
from sqlalchemy import create_engine,text
```

```python

# Replace with your actual database credentials
# Database credentials from the connection string
db_user = 'root'
db_password = 'your password'
db_host = 'localhost'
db_port = 3306
db_name = 'your database name'

# Construct the SQLAlchemy connection string without SSL CA
engine_url = f'mysql+pymysql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'

engine = create_engine(engine_url)

try:
    with engine.connect() as connection:
        for df in df_list:
            if not df.empty:
                # Assuming each DataFrame has a 'symbol' column
                ticker = df['symbol'].iloc[0].lower()
                table_name = ticker

                # Define the table creation SQL statement dynamically
                columns_sql = ', '.join([
                    f"`{col}` VARCHAR(255)" if df[col].dtype == 'object' else
                    f"`{col}` FLOAT" if df[col].dtype in ['float64', 'int64'] else
                    f"`{col}` DATETIME" if df[col].dtype == '<M8[ns]' else
                    f"`{col}` TEXT" for col in df.columns
                ])
                create_table_sql = f"""
                CREATE TABLE IF NOT EXISTS `{table_name}` (
                    {columns_sql},
                    INDEX (`date`)
                )
                """
                connection.execute(text(create_table_sql))

                # Insert data into the table
                df.to_sql(name=table_name, con=connection, if_exists='append', index=False)

    print("All data insertion complete.")

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # It's good practice to dispose of the engine when you're done
    engine.dispose()
```

```python
stock_volatility = {} 
for symbol_df in df_list:
    if not symbol_df.empty:
        symbol = symbol_df['symbol'].iloc[0]

        # Calculate daily return without modifying the original DataFrame
        daily_returns = []
        previous_close = None
        for close_price in symbol_df['close']:
            if previous_close is not None:
                daily_return = (close_price - previous_close) / previous_close
                daily_returns.append(daily_return)
            previous_close = close_price

        # Directly calculate volatility from the list
        volatility = pd.Series(daily_returns).std()
        stock_volatility[symbol] = volatility

# Get the top 10 most volatile stocks
if stock_volatility:
    sorted_volatility = sorted(stock_volatility.items(), key=lambda item: item[1], reverse=True)
    top_10_volatile = sorted_volatility[:10]

    # Create the DataFrame for the top 10 volatile stocks
    volatile_stocks_df = pd.DataFrame(top_10_volatile, columns=['Symbol', 'Volatility'])

    print("\nTop 10 Most Volatile Stocks (Standard Deviation of Daily Returns):")
    print(volatile_stocks_df)

    # Plotting the bar chart
    import matplotlib.pyplot as plt
    plt.figure(figsize=(12, 6))
    plt.bar(volatile_stocks_df['Symbol'], volatile_stocks_df['Volatility'], color='skyblue')
    plt.xlabel('Stock Symbol')
    plt.ylabel('Volatility (Standard Deviation of Daily Returns)')
    plt.title('Top 10 Most Volatile Stocks (Past Year)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("\nNo stock data available to calculate volatility.")
```

```python
cumulative_returns = {}
for symbol_df in df_list:
    if not symbol_df.empty:
        symbol = symbol_df['symbol'].iloc[0]

        # Calculate daily returns without modifying the original DataFrame
        daily_returns = []
        previous_close = None
        for close_price in symbol_df['close']:
            if previous_close is not None:
                daily_return = (close_price - previous_close) / previous_close
                daily_returns.append(daily_return)
            previous_close = close_price

        if daily_returns:  # Ensure at least one return exists
            # Compute cumulative return from daily return Series
            daily_return_series = pd.Series(daily_returns, index=symbol_df['date'].iloc[1:].values)
            cumulative_return_series = (1 + daily_return_series).cumprod() - 1
            cumulative_returns[symbol] = cumulative_return_series

# Identify the top 5 performing stocks based on the final cumulative return
if cumulative_returns:
    final_cumulative_returns = {
        symbol: returns.iloc[-1]
        for symbol, returns in cumulative_returns.items()
        if not returns.empty and pd.notnull(returns.iloc[-1])
    }
    top_5_performing = sorted(final_cumulative_returns.items(), key=lambda item: item[1], reverse=True)[:5]
    top_5_symbols = [symbol for symbol, _ in top_5_performing]

    print("\nTop 5 Performing Stocks (Based on Cumulative Return):")
    for symbol, return_val in top_5_performing:
        print(f"{symbol}: {return_val:.4f}")

    # Plotting cumulative returns for the top 5 stocks
    plt.figure(figsize=(14, 7))
    for symbol in top_5_symbols:
        if symbol in cumulative_returns:
            plt.plot(cumulative_returns[symbol].index, cumulative_returns[symbol].values, label=symbol)

    plt.xlabel('Date')
    plt.ylabel('Cumulative Return')
    plt.title('Cumulative Return Over Time for Top 5 Performing Stocks')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("\nNo stock data available to calculate cumulative returns.")
```

```python
def get_monthly_prices_raw(price_series):
    """
    Gets the last available price for each month in a price series without using resample().

    Args:
        price_series (pd.Series): A pandas Series of stock prices indexed by date.

    Returns:
        pd.Series: A pandas Series of monthly prices indexed by the end of each month.
    """
    monthly_prices = {}
    for date, price in price_series.items():
        year_month = date.to_period('M')
        monthly_prices[year_month] = price
    return pd.Series(monthly_prices)

def calculate_monthly_return_raw(monthly_price_series):
    """
    Calculates the monthly return from a series of monthly prices without using pct_change().

    Args:
        monthly_price_series (pd.Series): A pandas Series of monthly prices.

    Returns:
        pd.Series: A pandas Series of monthly returns.
    """
    monthly_returns = []
    previous_month_price = None
    monthly_dates = []
    sorted_monthly_prices = monthly_price_series.sort_index()

    for month, current_month_price in sorted_monthly_prices.items():
        if previous_month_price is not None:
            monthly_return = (current_month_price - previous_month_price) / previous_month_price
            monthly_returns.append(monthly_return)
            monthly_dates.append(month.to_timestamp('M'))
        previous_month_price = current_month_price

    return pd.Series(monthly_returns, index=monthly_dates)

# Assuming df_list is already populated with your stock data DataFrames

monthly_returns_data = {}
for symbol_df in df_list:
    if not symbol_df.empty:
        symbol = symbol_df['symbol'].iloc[0]

        # Ensure 'date' or 'Date' is datetime
        date_column = None
        if 'date' in symbol_df.columns:
            date_column = 'date'
        elif 'Date' in symbol_df.columns:
            date_column = 'Date'

        if date_column is not None:
            symbol_df[date_column] = pd.to_datetime(symbol_df[date_column])
            symbol_df_with_date = symbol_df.set_index(date_column) # Keep the original date column

            monthly_prices = get_monthly_prices_raw(symbol_df_with_date['close'])
            monthly_returns_data[symbol] = calculate_monthly_return_raw(monthly_prices)
        else:
            print(f"Warning: No 'date' or 'Date' column found for {symbol}. Skipping monthly return calculation.")
            continue

# Identify top 5 gainers and losers for each month
monthly_top_gainers_losers = {}
all_months = sorted(list(set(month for returns in monthly_returns_data.values() for month in returns.index)))
num_months = len(all_months)

# Calculate the number of rows needed for 4 plots per row
num_rows = (num_months + 3) // 4 * 2  # Two rows for each month (gainers and losers)

fig, axes = plt.subplots(nrows=num_rows, ncols=4, figsize=(24, 6 * num_rows))
axes = axes.flatten()

plot_index = 0
for i, month in enumerate(all_months):
    month_data = []
    for symbol, returns in monthly_returns_data.items():
        if month in returns:
            month_data.append({'symbol': symbol, 'return': returns[month]})

    if month_data:
        top_n = sorted(month_data, key=lambda x: x['return'], reverse=True)[:5]
        bottom_n = sorted(month_data, key=lambda x: x['return'])[:5]

        print(f"\n----- {month.strftime('%Y-%m')} -----")
        print("Top 5 Gainers:")
        for gainer in top_n:
            print(f"{gainer['symbol']}: {gainer['return']:.4f}")
        print("\nTop 5 Losers:")
        for loser in bottom_n:
            print(f"{loser['symbol']}: {loser['return']:.4f}")

        # Plot top 5 gainers
        ax_gainers = axes[plot_index]
        symbols_gainers = [item['symbol'] for item in top_n]
        returns_gainers = [item['return'] * 100 for item in top_n]
        ax_gainers.bar(symbols_gainers, returns_gainers, color='green')
        ax_gainers.set_title(f'Top 5 Gainers - {month.strftime("%Y-%m")}', fontsize=10)
        ax_gainers.set_ylabel('Monthly Return (%)', fontsize=8)
        ax_gainers.set_xticks(range(len(symbols_gainers)))
        ax_gainers.set_xticklabels(symbols_gainers, rotation=45, ha='right', fontsize=8)
        ax_gainers.tick_params(axis='y', labelsize=8)

        plot_index += 1

        # Plot top 5 losers
        ax_losers = axes[plot_index]
        symbols_losers = [item['symbol'] for item in bottom_n]
        returns_losers = [item['return'] * 100 for item in bottom_n]
        ax_losers.bar(symbols_losers, returns_losers, color='red')
        ax_losers.set_title(f'Top 5 Losers - {month.strftime("%Y-%m")}', fontsize=10)
        ax_losers.set_ylabel('Monthly Return (%)', fontsize=8)
        ax_losers.set_xticks(range(len(symbols_losers)))
        ax_losers.set_xticklabels(symbols_losers, rotation=45, ha='right', fontsize=8)
        ax_losers.tick_params(axis='y', labelsize=8)

        plot_index += 1
    else:
        # Handle cases with no data for a month by turning off the corresponding subplots
        axes[plot_index].axis('off')
        plot_index += 1
        axes[plot_index].axis('off')
        plot_index += 1

# Remove any unused subplots
if plot_index < len(axes):
    for i in range(plot_index, len(axes)):
        fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
